```python
def create_data(features):
    length = dataset.shape[0]
    x = np.ones((length, 1))
    for i in features:
        x1 = np.array(dataset[i]).reshape(length, 1)
        x = np.concatenate((x, x1), axis=1)
    y = np.array(dataset.Chance_of_Admit_).reshape(length, 1)
    return x, y
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def cost(x, y, theta):
    h_x = np.dot(x, theta)
    prob = sigmoid(h_x)
    n = x.shape[0]
    total_cost = -(1 / n) * np.sum(y * np.log(prob) + (1 - y) * np.log(1 - prob))
    return total_cost
```

# Gradient Descent

```python
def gradient_descent(x, y, theta, lr, iterations_num):
    pbar = ProgressBar()
    loss_history = []
    t_hist = []
    for i in pbar(range(iterations_num)):
        h_x = np.dot(x, theta)
        prob = sigmoid(h_x)
        n = x.shape[0]
        cost_ = cost(x, y, theta)
        loss_history.append(cost_)
        t_hist.append(theta)
        gradient = np.dot(x.T,  prob - y)
        gradient /= n
        gradient *= lr
        theta -= gradient
    return theta, loss_history, t_hist
```
